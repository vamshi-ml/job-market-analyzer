```python
 # ### **TASK 3: Calculate Key Metrics**

#**Question 1:** What's the average salary for data science jobs?

Datascience_Avg_salary = data_cleaned['salary_in_usd'].mean()
print(f'{Datascience_Avg_salary:.2f}')
print("-" * 40)

#Jobs per experience level?
print(data_cleaned.experience_level.value_counts())
print("-" * 40)

#Most common job title?

print(data_cleaned.job_title.value_counts().head().index[0])

print("-" * 40)

#Average salary by experience?

print(data_cleaned.groupby('experience_level')['salary_in_usd'].mean())

print("-" * 40)

#Top 5 countries with most jobs?
data_cleaned.company_location.value_counts().head(5)
```

```python
# ### **TASK 4: Create Visualizations**

# **Chart 1:** Bar chart showing number of jobs by experience level

import matplotlib.pyplot as plt
Exp_level=data_cleaned['experience_level'].value_counts()
colors =['Blue','Orange','Pink','Green']

plt.bar(Exp_level.index,Exp_level.values,color=colors)
for i in range(len(Exp_level)):
  plt.text(i,Exp_level.values[i],Exp_level.values[i],ha='center')
plt.xlabel("experience_level")
plt.ylabel(" Number of jobs")
plt.title("number of jobs by experience level")


#**Chart 2:** Bar chart showing average salary by job title (top 10 titles)
avg_salary=data_cleaned.groupby('job_title')['salary_in_usd'].mean().head(10)
avg_salary_sorted=avg_salary.sort_index(ascending=True)
top_10_jobs = avg_salary_sorted
#plt.barh(top_10_jobs.index,top_10_jobs.values,color = plt.cm.tab10(range(len(top_10_jobs))))
plt.figure(figsize=(10,6))

plt.barh(
    top_10_jobs.index,
    top_10_jobs.values,
    color=plt.cm.tab10(range(len(top_10_jobs)))
)

plt.xlabel("Average Salary (USD)")
plt.ylabel("Job Title")
plt.title("Top 10 Job Titles by Average Salary")
plt.show()


#**Chart 3:** Bar chart showing top 10 countries with most jobs
most_job_countries=data_cleaned.company_location.value_counts().head(10)
plt.barh(most_job_countries.index,most_job_countries.values, color=plt.cm.tab10(range(len(most_job_countries))))
plt.xlabel("most_job_countries")
plt.ylabel("Count")
plt.title("Bar chart showing top 10 countries with most jobs")
plt.show()
```
